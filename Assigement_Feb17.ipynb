Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

MongoDB is basically document based database(JSON like document). It ia No-SQL database.It was designed to handle large volumes of unstructured and semi-structured data.

Non-relational databases, also known as NoSQL databases, are a type of database management system that does not use the traditional relational database model. Instead, they use different data models, such as key-value, document-oriented, graph, or column-family databases. Non-relational databases are often preferred for their ability to handle large amounts of data, scale horizontally, and provide flexibility in data modeling.

MongoDB is often preferred over SQL databases in scenarios where there is a need for high performance, scalability, and flexibility in data modeling. Some specific scenarios where MongoDB is preferred over SQL databases include:

1.Big Data: MongoDB is particularly suited for handling large volumes of unstructured or semi-structured data, making it a popular choice for big data applications.

2.Agile development: MongoDB's flexible schema design allows for easier and faster iteration during the development process.

3.Real-time analytics: MongoDB can handle real-time analytics with ease, making it a popular choice for applications that require fast data processing.

4.Cloud-based applications: MongoDB is a good fit for cloud-based applications that require easy scalability, high availability, and redundancy.

5.Mobile and web applications: MongoDB's ability to handle unstructured data makes it a good fit for mobile and web applications, which often deal with data in various formats.

Overall, MongoDB is a good choice for applications that require scalability, high performance, and flexibility in data modeling.





Q2. State and Explain the features of MongoDB.

MongoDB is a document-oriented NoSQL database that offers a variety of features for flexible, scalable, and high-performance data storage. Here are some of the key features of MongoDB:

1.Document-oriented: MongoDB stores data in flexible JSON-like documents that can vary in structure and content, allowing for easy scalability and data modeling.

2.Dynamic schema: MongoDB's dynamic schema allows for quick and easy modifications to the database structure without downtime or complex migration procedures.

3.Indexing: MongoDB supports various types of indexing, including single-field, compound, and multi-key indexes, for faster queries and data retrieval.

4.Replication and high availability: MongoDB's replication feature ensures that data is automatically replicated to multiple nodes for high availability and fault tolerance.

5.Sharding: MongoDB's sharding feature allows for horizontal scaling of data across multiple nodes, enabling large datasets to be distributed across multiple servers.

6.Aggregation framework: MongoDB's powerful aggregation framework allows for complex data analysis and processing, such as data filtering, grouping, sorting, and joining.

7.GridFS: MongoDB's GridFS allows for the storage and retrieval of large files and media, such as images, audio, and video.

8.Security: MongoDB offers various security features, such as authentication, authorization, encryption, and auditing, to protect data from unauthorized access and ensure compliance with data protection regulations.

9.Integration with programming languages: MongoDB offers drivers and APIs for various programming languages, including Python, Java, Node.js, and Ruby, making it easy to integrate with different software environments.

Overall, MongoDB offers a wide range of features that make it a popular choice for modern, data-intensive applications that require flexible, scalable, and high-performance data storage.

Q6.Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

In MongoDB, the sort() method is used to sort the results of a query in ascending or descending order based on one or more fields. The sort() method takes a dictionary of field names and sort directions as its argument, where the keys are field names and the values are either 1 (ascending order) or -1 (descending order).

Q7.Explain why delete_one(), delete_many(), and drop() is used.

1.delete_one() :- is used to remove a single document from a collection that matches a given query criteria. If multiple documents match the criteria, only the first matching document will be deleted.

2.delete_many() :- is used to remove multiple documents from a collection that match a given query criteria. This method deletes all documents that match the criteria, not just the first one.

3.drop() :- is used to remove an entire collection from a database. This method permanently deletes all documents in the collection and removes the collection from the database.
